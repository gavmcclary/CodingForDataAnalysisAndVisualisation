R = 0 --> Not linearly correlated

R = -1 --> Strong negative relationship

# Spearman's Rank Correlation

R = 1 --> Strong positive relationships

R = 0 --> Not correlated

R = -1 --> Strong negative relationship

```python
from scipy.stats import spearmanr
rcParams['figure.figsize'] = 14,7
```

```python
cars.head()
```

```python
sb.pairplot(cars)
```

```python
x = cars[['cyl', 'vs', 'am', 'gear']]
sb.pairplot(x)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coefficient, p_value = spearmanr(cyl, vs)

print('Spearman Rank Correlation Coeficient %0.3f' % (spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, am)

print('Spearman Rank Correlation Coeficient %0.3f' % (spearmanr_coefficient))
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, gear)

print('Spearman Rank Correlation Coeficient %0.3f' % (spearmanr_coefficient))
```

# Chi-square test for independence

```python
table = pd.crosstab(cyl, am)

from scipy.stats import chi2_contingency
```

```python
chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl, vs)

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

```python
table = pd.crosstab(cyl, gear)

chi2, p, dof, expected = chi2_contingency(table.values)
print('Chi-square statistic %0.3f p_value %0.3f' % (chi2, p))
```

# Scaling and Transforming

Normalisation: purrint each obervation on a relative scale

between the values of 0 and 1

            Value of observation
            --------------------
      Sum of all observations in variable
            
Standardization: rescaling data so that it has a zero mean and 

unit variance

```python
import sklearn
from sklearn import preprocessing
from sklearn.preprocessing import scale
rcParams['figure.figsize'] = 5,4
sb.set_style('whitegrid')
```

```python
mpg = cars.mpg
plt.plot(mpg)
```

```python
cars[['mpg']].describe()
```

```python
mpg_matrix = mpg.values.reshape(-1,1)
```

```python
scaled = preprocessing.MinMaxScaler()

scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
scaled = preprocessing.MinMaxScaler(feature_range=(0,10))

scaled_mpg = scaled.fit_transform(mpg_matrix)
plt.plot(scaled_mpg)
```

```python
standarized_mpg = scale(mpg, axis=0, with_mean=False, with_std=False)
```

```python
plt.plot(standarized_mpg)
```
